```python
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen as uReq
import re

baseurl='https://www.thewhiskyexchange.com/'
#print(baseurl)

headers={
    'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'
}

#r=requests.get(baseurl)
productlinks=[]

for x in range(1,3):
    
    #url1='https://www.thewhiskyexchange.com/c/35/japanese-whisky'
    url=f'https://www.thewhiskyexchange.com/c/35/japanese-whisky?pg={x}#productlist-filter'
    #print(url)
    r=requests.get(url)
    #print(r)
    soup=BeautifulSoup(r.content,'html.parser')
    #soup1=BeautifulSoup(r.content,'lxml') we can use both 'html ya lxml' 
    #print(soup)
    productlist = soup.find_all('div',class_='group group-list')
    #print(productlist)
    for item in productlist:
        for link in item.find_all('a',href=True):
            #print(link['href']) # for checking 
            #print(link)
            productlinks.append(baseurl + link['href'])
#print('\n',productlinks,'\n')

whiskylist=[]

for links in productlinks:
#testlink='https://www.thewhiskyexchange.com/p/2947/nikka-yoichi-20-year-old'

    r = requests.get(links, headers=headers)
    soup = BeautifulSoup(r.content,'html.parser')
    name = soup.find('h1',class_='product-main__name').text.strip()
    alcohol_data=soup.find('p',class_='product-main__data').text
    
    #rating=soup.find('span',class_='review-overview__rating star-rating star-rating--50').text.strip()
    
    stores=soup.find('p',class_='product-action__stock-flag').text.strip()
    price=soup.find('p',class_='product-action__price').text.strip()
    vat_price=soup.find('p',class_='product-action__vat-price').text.strip()
    per_liter_price=soup.find('p',class_='product-action__unit-price').text.strip()
    #print(per_liter_price)
    product_des = soup.findAll('div',attrs={"class":"product-main__description"})
 
        
    try:
        rating=soup.find('span',class_='review-overview__rating star-rating star-rating--50').text.strip()
        offer=soup.find('span',class_='product-offer__text').text.strip()
        product_message=soup.find('p',class_='product-action__message').text.strip()
        product_name=soup.find('ul',class_='product-main__meta').text.strip()
        review=soup.find('span',class_='review-overview__count').text.strip()
        for x in product_des:
            product_description=x.find('p').text.strip()
    
    except:
        rating=""
        offer=""
        product_message=""
        product_name=""
        review=""
        product_description=""
    whisky = {
        'name':name,
        'product_name':product_name,
        'alcohol_data':alcohol_data,
        'rating':rating,
        'review':review,
        'offer':offer,
        'stores':stores,
        'price':price,
        'vat_price':vat_price,
        'per_liter_price':per_liter_price,
        'product_message':product_message,
        'product_description':product_description
        }
    whiskylist.append(whisky)
print(len(whiskylist))
import pandas as pd
df=pd.DataFrame(whiskylist)
#df.head
df.to_csv("japenese_whisky.csv")

```
